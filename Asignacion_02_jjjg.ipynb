#Algoritmo para encontrar el número Máximo

```python
#Partes de un número
#Si es par, lo divides entre 2
#Si es impar lo multiplicas por 3 y le sumas 1
#Siempre terminas en la secuencia 4 2 1
```

```python
import matplotlib.pyplot as plt
```

```python
def es_par(numero):
  if numero%2==0:
    return True
  else:
    return False

#x = El número de secuencias que habrá
v1=[]
v2=0
n_max=0
for x in range(1,1001,1):
  contador=1
  inicio=x
  n=inicio
  i=1
  iteracion=[i]
  secuencia=[n]
  
  continuar = True
  while continuar:
    if es_par(n):
      n=n/2
    else:
      n=(3*n)+1
    i+=1
    contador=contador+1
    iteracion.append(i)
    secuencia.append(n)
    if n==1:
      continuar=False
  plt.plot(iteracion,secuencia,marker="o")
  n_max=max(secuencia)
  v1.append(n_max) #vector que guarda el número máximo de cada secuencia
  
  print("Iteración número {}".format(x))
  print("El número de puntos graficados es: {}".format(contador))
  print("La secuencia es: ")
  print(secuencia)
  print("El número máximo es: {}".format(n_max))


  x1=secuencia.index(n_max)
  y1=n_max

  #print("El número máximo se encuentra en ({},{})".format(x1+1,y1) )
  print("El número máximo se encuentra en el índice {} de la iteración {}".format(x1+1,x))
  print("--------------------")

print("\n*****************************")
print("El arreglo de números máximos es: ")
print(v1)
v2=max(v1) #Obtiene el número máximo del arreglo de números máximos
x2=x1=v1.index(v2)
print("\nEl número máximo de todas las secuencias es: {} y se encuentra en la iteración {}".format(v2,x2+1))
```

```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs,make_moons,make_circles
import seaborn as sns
import pandas as pd
```

#Generar datos aleatorios con una distribución normal


```python
np.random.seed(0)
media = 1
std = 2
n=400
x = np.random.normal(media,std,(n,1))
y = np.random.normal(3*media,3*std,(n,1))
print(x.shape,y.shape)
plt.scatter(x,y)
```

```python
#Creamos unas listas para los datos y así obtener sus máximos y mínimos
#Posteriormente se crea un dataframe para usar la biblioteca sns
x_list =[]
y_list =[]

for i in range (n):
  x_list.append(float(x[i]))
  y_list.append(float(y[i]))

x_max = max(x_list)
y_max = max(y_list)
x_min = min(x_list)
y_min = min(y_list)
```

```python
df = pd.DataFrame()

df['x_data'] = x_list
df['y_data'] = y_list

print(df)
print("El máximo de x es: {}, el mínimo de x es: {}".format(x_max,x_min))
print("El máximo de Y es: {}, el mínimo de y es: {} ".format(y_max,y_min))
```

```python
sns.set_theme(style="darkgrid")
g = sns.jointplot(x="x_data", y="y_data", data=df,
                  kind="reg", truncate=False,
                  xlim=(-6, 6), ylim=(-20, 20),
                  color="m", height=7)
```

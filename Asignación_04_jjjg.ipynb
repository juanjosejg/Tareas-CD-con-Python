```python
#Asignación_04_JJJG
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_aleatorio = 0
def tirar_dado():
  num_aleatorio = np.random.randint(1,7)
  return num_aleatorio

vector_aleatorio=[]

for i in range(1000):
  num = tirar_dado()
  vector_aleatorio.append(num)
```

```python
intervalos = range(1,8)
#        datos, intervalos, color, separación
plt.hist(x=vector_aleatorio, bins=intervalos, color='#1b5ed6', rwidth=0.8)
plt.title('Histograma de tiros')
plt.xlabel('Tiros')
plt.ylabel('Frecuencia')
```

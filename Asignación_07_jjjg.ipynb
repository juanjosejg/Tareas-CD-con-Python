```python
lista=np.array([13,8,46,61,28,7,93,10,7,100,4,16,23,33,18,51,26,19,20,54,19,37,13,8,28,25,4,12,12])
```

```python
lista_ordenada=np.sort(lista)

promedio=lista.mean()
desv=lista.std()
mediana=np.median(lista)
p25=np.percentile(lista, 25)
p50=np.percentile(lista, 50)
p75=np.percentile(lista, 75)
```

```python
plt.hist(lista)
```

```python
sns.boxplot(lista)
```

# Resultados

```python
print("El promedio de los datos es: ",promedio)
print("\nLa desviación estándar de los datos es: ",desv)
print("\nLos percentiles 25, 50 y 75 son: ",p25,",",p50,",",p75 ,"respectivamente")
print("""\nCon base en las gráficas de la sección de cálculos se puede determinar 
que los datos no representan una distribución normal """)
```

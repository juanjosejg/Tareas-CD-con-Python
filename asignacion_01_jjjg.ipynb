```python
#Partes de un número
#Si es par, lo divides entre 2
#Si es impar lo multiplicas por 3 y le sumas 1
#Siempre terminas en la secuencia 4 2 1
```

```python
import matplotlib.pyplot as plt
```

```python
def es_par(numero):
  if numero%2==0:
    return True
  else:
    return False

#x = El número de secuencias que habrá
for x in range(1,101,1):
  inicio=x
  n=inicio
  i=1
  iteracion=[i]
  secuencia=[n]
  continuar = True

  while continuar:
    if es_par(n):
      n=n/2
    else:
      n=(3*n)+1
    i+=1
    iteracion.append(i)
    secuencia.append(n)
    if n==1:
      continuar=False

  plt.plot(iteracion,secuencia,marker="o")
```
